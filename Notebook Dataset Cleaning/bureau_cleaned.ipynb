```python
dfe = df.copy()
# Memeriksa baris duplikat
duplicate_rows = dfe.duplicated()

# Menampilkan jumlah baris duplikat
print(f"Jumlah baris duplikat: {duplicate_rows.sum()}")

# Fungsi untuk menghitung dan menampilkan kolom dengan nilai missing
def check_nan(dfe):
    # Hitung nilai null
    null_values = dfe.isnull().sum()

    # Buat DataFrame untuk total nilai null dan persentase
    nvc = pd.DataFrame(null_values[null_values > 0].sort_values(), columns=['Total Null Values'])  # Hanya ambil kolom dengan nilai null
    nvc['Percentage'] = (nvc['Total Null Values'] / dfe.shape[0]) * 100
    nvc["Data Type"] = [dfe[col].dtype for col in nvc.index]  # Ambil tipe data dari kolom yang ada

    # Urutkan berdasarkan total nilai null dan persentase
    nvc.sort_values(by=["Total Null Values", "Percentage"], ascending=False, inplace=True)

    # Tampilkan hasil dengan gaya
    return nvc.style.background_gradient(cmap='Blues')

# Panggil fungsi dan simpan hasilnya
styled_result = check_nan(dfe)

# Untuk menampilkan hasil di Jupyter Notebook, gunakan:
styled_result
```

```python
dfe = df.copy()
# Memeriksa baris duplikat
duplicate_rows = dfe.duplicated()

# Menampilkan jumlah baris duplikat
print(f"Jumlah baris duplikat: {duplicate_rows.sum()}")

# Fungsi untuk menghitung dan menampilkan kolom dengan nilai missing
def check_nan(dfe):
    # Hitung nilai null
    null_values = dfe.isnull().sum()

    # Buat DataFrame untuk total nilai null dan persentase
    nvc = pd.DataFrame(null_values[null_values > 0].sort_values(), columns=['Total Null Values'])  # Hanya ambil kolom dengan nilai null
    nvc['Percentage'] = (nvc['Total Null Values'] / dfe.shape[0]) * 100
    nvc["Data Type"] = [dfe[col].dtype for col in nvc.index]  # Ambil tipe data dari kolom yang ada

    # Urutkan berdasarkan total nilai null dan persentase
    nvc.sort_values(by=["Total Null Values", "Percentage"], ascending=False, inplace=True)

    # Tampilkan hasil dengan gaya
    return nvc.style.background_gradient(cmap='Blues')

# Panggil fungsi dan simpan hasilnya
styled_result = check_nan(dfe)

# Untuk menampilkan hasil di Jupyter Notebook, gunakan:
styled_result
```

```python
nan_info = dfe.isnull().sum()
nan_percent = nan_info / len(dfe) * 100
missing = nan_percent[nan_percent > 0]

#1. Penanganan kolom dengan missing value lebih dari 50% Drop kolom
# Jika lebih dari setengah data kosong, kemungkinan besar kolom tidak informatif dan bisa membahayakan model
dfe = dfe.drop(columns=missing[missing > 50].index)

# 2. Penanganan kolom dengan missing value antara 5% - 50%  Imputasi (pengisian)
# Masih cukup banyak data yang tersedia, jadi kita bisa isi dengan nilai yang representatif
from sklearn.impute import SimpleImputer
import numpy as np

# Tentukan kolom yang akan diimputasi
to_impute = missing[(missing > 5) & (missing <= 50)].index

# Imputasi median (lebih aman untuk data skewed)
imputer = SimpleImputer(strategy='median')
dfe[to_impute] = imputer.fit_transform(dfe[to_impute])

# 3. Penanganan kolom dengan missing value < 5% Imputasi atau Drop baris
# Jumlah baris yang akan dibuang sedikit, jadi aman jika kita drop saja
low_missing = missing[missing <= 5].index
dfe = dfe.dropna(subset=low_missing)
```

```python
# Panggil fungsi dan simpan hasilnya
styled_result = check_nan(dfe)

# Untuk menampilkan hasil di Jupyter Notebook, gunakan:
styled_result
```
