```python
dfe = df.copy()
# Memeriksa baris duplikat
duplicate_rows = dfe.duplicated()

# Menampilkan jumlah baris duplikat
print(f"Jumlah baris duplikat: {duplicate_rows.sum()}")

# Fungsi untuk menghitung dan menampilkan kolom dengan nilai missing
def check_nan(dfe):
    # Hitung nilai null
    null_values = dfe.isnull().sum()

    # Buat DataFrame untuk total nilai null dan persentase
    nvc = pd.DataFrame(null_values[null_values > 0].sort_values(), columns=['Total Null Values'])  # Hanya ambil kolom dengan nilai null
    nvc['Percentage'] = (nvc['Total Null Values'] / dfe.shape[0]) * 100
    nvc["Data Type"] = [dfe[col].dtype for col in nvc.index]  # Ambil tipe data dari kolom yang ada

    # Urutkan berdasarkan total nilai null dan persentase
    nvc.sort_values(by=["Total Null Values", "Percentage"], ascending=False, inplace=True)

    # Tampilkan hasil dengan gaya
    return nvc.style.background_gradient(cmap='Blues')

# Panggil fungsi dan simpan hasilnya
styled_result = check_nan(dfe)

# Untuk menampilkan hasil di Jupyter Notebook, gunakan:
styled_result
```

```python
# 1. Drop kolom dengan missing value > 50%
missing_threshold = 0.5
cols_to_drop = dfe.columns[dfe.isnull().mean() > missing_threshold]
dfe.drop(columns=cols_to_drop, inplace=True)

# 2. Imputasi kolom numerik dengan missing value antara 10% - 50% menggunakan median
num_cols = dfe.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    missing_pct = dfe[col].isnull().mean()
    if 0.1 < missing_pct <= 0.5:
        dfe[col].fillna(dfe[col].median(), inplace=True)

# 3. Imputasi kolom kategorikal dengan missing value antara 10% - 50% menggunakan modus
cat_cols = dfe.select_dtypes(include='object').columns
for col in cat_cols:
    missing_pct = dfe[col].isnull().mean()
    if 0.1 < missing_pct <= 0.5:
        dfe[col].fillna(dfe[col].mode()[0], inplace=True)

# 4. Imputasi kolom dengan missing value < 10%
for col in dfe.columns:
    if dfe[col].isnull().mean() <= 0.1:
        if dfe[col].dtype == 'object':
            dfe[col].fillna(dfe[col].mode()[0], inplace=True)
        else:
            dfe[col].fillna(dfe[col].median(), inplace=True)

# 5. Cek hasil akhir
# Panggil fungsi dan simpan hasilnya
styled_result = check_nan(dfe)

# Untuk menampilkan hasil di Jupyter Notebook, gunakan:
styled_result
```

1. Drop kolom dengan missing value > 50%, Karena lebih dari separuh datanya hilang, kolom tersebut kemungkinan besar tidak lagi memberikan informasi yang cukup untuk dianalisis.
   
2. Imputasi numerik 10–50% dengan median, Median tidak terpengaruh outlier sehingga lebih aman digunakan untuk mengisi nilai kosong pada distribusi data numerik yang tidak simetris.
   
3. Imputasi kategorikal 10–50% dengan modus, menggunakan modus menjaga konsistensi pola kategori terbanyak tanpa mengubah distribusi asli secara signifikan.

4: Imputasi <10% dengan median/modus, Persentase kecil dapat diisi langsung tanpa risiko distorsi besar terhadap struktur data secara keseluruhan.
